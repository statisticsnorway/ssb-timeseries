```python
import pandas as pd
```

```python
data1 = {
    "periode": ['2023-01', '2023-02', '2023-03'],
    "verdi": [234, 543, 643],
    "user": ['1', '1', '1']
}

data2 = {
    "periode": ['2023-02', '2023-03', '2023-04'],
    "verdi": [543, 650, 712],
    "user": ['2', '2', '2']
}
```

```python
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
map_periode = df2.periode.isin(df1.periode)
map_verdi = df2.verdi.isin(df1.verdi)
map_isin = map_periode & map_verdi
map_isnotin = ~map_isin
```

```python
df2[map_isin]
```

```python
df2[map_isnotin]
```
